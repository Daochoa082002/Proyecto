# Grafica de Barras

Para graficar un Barplot usamos las razones de educacion de mujeres frente a hombres, en la primaria, secundaria y terciaria. 

```python
df_estudios = df_proyecto[df_proyecto["Indicator Name"] == "Razon de educacion de mujeres frente a hombres"]
```

```python
df_estudios = df_estudios.rename(columns = {"Sexo_o_clase" : "Estudios"})
```

```python
df_estudios = df_estudios[df_estudios["Income_Group"] != "No clasificado"]
```

```python
df_estudios = df_estudios[df_estudios["Estudios"] != "Primaria y Secundaria"]
```

```python
df_tabla_est = pd.pivot_table(df_estudios, 
                                values=['2000',"2001","2002","2003","2004","2005","2006","2007","2008","2009",
                                       "2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020"], 
                                index=["Indicator Name"],
                                    columns=['Country Name',"Estudios","Income_Group","Region"], 
                                aggfunc=np.mean)
df_tabla_est = pd.DataFrame(df_tabla_est)
```

```python
df_tabla_est = df_tabla_est.transpose().reset_index()
```

```python
df_tabla_est.head()
```

```python
order = ["Ingreso bajo","Ingreso mediano bajo","Ingreso mediano alto","Ingreso alto"]
```

```python
ax = plt.subplots()
sns.set_palette('Set1')
sns.barplot(data = df_tabla_est,
            estimator = np.mean,
            x = "Razon de educacion de mujeres frente a hombres",
            y = "Income_Group",
            hue = "Estudios",
            order = order,
            ci = None,
            orient = "h")
plt.title('Educacion de mujeres frente a hombres en los paises segun su clasificacion de ingreso')
plt.xlabel('Razon')
plt.ylabel('',size=1)
plt.legend(ncol=1, bbox_to_anchor = (1, 0.9))
sns.despine()
plt.show()
```

## Analisis:
Las razon se halla mediante una operación en la que se divide las tasas de mujeres sobre la de hombres, en la que se puede concluir que si el resultado es igual a 1 la razon es equilibrada para ambos sexos y si es menor o mayor hay una desigualdad respecto a las variables.
